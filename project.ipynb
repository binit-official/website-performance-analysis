```python
 
 import numpy as np
 import matplotlib.pyplot as plt
 import pandas as pd
 import seaborn as sns
```

```python
df=pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df=df.drop(index=0).reset_index(drop=True)
df.columns=["channel group","DateHour","Users","Sessions","Engaged Session","Average engagement time per session",	"Engaged sessions per user",	"Events per session",	"Engagement rate",	"Event count"]
```

```python
df.head()

```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors='coerce')
```

```python

df.info()
```

```python
df.head()
```

```python
# Select all columns except "channel group" and "DateHour"
numeric_cols = [col for col in df.columns if col not in ["channel group", "DateHour"]]

# Convert them to numeric (coerce errors to NaN)
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Extract hour from datetime column
df["Hour"] = df["DateHour"].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# SEssions and users over time
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users Over Time")
plt.xlabel("DateHour")
plt.ylabel("Count")
```

```python
# Plotting the distribution of "Users" by "channel group"
```

# Plotting the distribution of "Users" by "channel group"

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("Total Users by Channel Group")
plt.xticks(rotation=45)
plt.xlabel("Channel Group")
plt.ylabel("Total Users")
plt.show()
```

```python
df.head()
```

Average engagement time by channel


```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group", y="Average engagement time per session", estimator=np.mean, palette="magma")
plt.title("Average Engagement Time by Channel Group")
plt.xticks(rotation=45)
plt.xlabel("Channel Group")
plt.ylabel("Average Engagement Time (seconds)")
plt.show()
```

```python
df.head()
```

```python
# Engagement rate Distribution by channel 
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="channel group", y="Engagement rate", data=df, palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel Group")
plt.xticks(rotation=45)
plt.xlabel("Channel Group")
plt.ylabel("Engagement Rate")
plt.show()
```

```python
# Engaged ns non engaged sessions 
```

```python
session_df= df.groupby("channel group")[["Sessions","Engaged Session"]].sum().reset_index()
session_df["Non Engaged Sessions"] = session_df["Sessions"] - session_df["Engaged Session"]
session_df_melted = session_df.melt(id_vars="channel group",
                                    value_vars=["Engaged Session", "Non Engaged Sessions"])
plt.figure(figsize=(8, 5))
sns.barplot(data=session_df_melted, x="channel group", y="value", hue="variable")
plt.title("Engaged vs Non-Engaged Sessions by Channel Group")
plt.xticks(rotation=45)
plt.xlabel("Channel Group")
plt.ylabel("Number of Sessions")
plt.show()
```

```python
# Traffic by hour and channel
```

```python
heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.title("Traffic by Hour and Channel Group")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

```python
# Engagement rate vs Session over time
```

```python
df_plot= df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement Rate and Sessions Over Time")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.show()
```
